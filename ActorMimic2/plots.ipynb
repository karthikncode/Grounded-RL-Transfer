```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_vals(seed, trf):
    loc = './logs/fixedpadded2_transfer_seed_' + seed + '_t_' + trf + '/1/test_avgR.log'  
    f1 = open(loc)
    l1 = f1.readlines()
    l2 = ''.join(l1).split()
    print l2[:3]
    l3 = map(float, l2[2:])
    return l3
```

```python
dic = {}
for seed in ['0', '3', '5', '7']:
    for trf in ['0', '1', '2']:
       dic[seed + '_' + trf] = get_vals(seed, trf) 
```

```python
seeds = ['0','3','5','7']
trfs = ['0', '1', '2']
names = ['multitask_transfer', 'singletask', 'multitask']
colors = [plt.cm.Reds, plt.cm.Greens, plt.cm.Blues]
base_colors = ['r', 'g', 'b']
```

```python
import numpy as np
```

```python
def plot(trf):
    plt.grid()
    plt.ylim(0,45)
    print trf
    colormap = colors[int(trf)]
    num_plots = len(seeds) + 1
    plt.gca().set_color_cycle([colormap(i) for i in np.linspace(0, 0.9, 10)])
    avg = np.zeros(360)
    for seed in seeds:
        l3 = dic[seed + '_' + trf]
        print len(l3)
        avg = avg + np.array(l3[:360])
        plt.plot(l3)
    avg = avg / len(seeds)
    plt.plot(avg, color = base_colors[int(trf)], label = names[int(trf)])
    plt.xlabel('Steps / 5000')
    plt.ylabel('Average Reward')
    plt.legend(fontsize = 'x-small')
```

```python
def plotall():
    plt.grid()
    plt.ylim(0,45)
    for trf in trfs:
        colormap = colors[int(trf)]
        num_plots = len(seeds) + 1
        plt.gca().set_color_cycle([colormap(i) for i in np.linspace(0, 0.9, 10)])
        avg = np.zeros(360)
        for seed in seeds:
            l3 = dic[seed + '_' + trf]
            print len(l3)
            avg = avg + np.array(l3[:360])
            #plt.plot(l3)
        avg = avg / len(seeds)
        plt.plot(avg, color = base_colors[int(trf)], label = names[int(trf)])
        plt.xlabel('Steps / 5000')
        plt.ylabel('Average Reward')
        plt.legend(fontsize = 'x-small')
```

```python
plotall()
```
